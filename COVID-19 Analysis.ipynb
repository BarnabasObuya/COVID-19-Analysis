```python
# Create a bar plot for 'Population' data
ax1 = sns.barplot(data=data, y='Population', x='Continent', ci=True, color='blue')
ax1.set_ylabel('Population', color='blue')
ax1.set_xlabel('Continent')
ax1.set_title('Population and Total Cases by Continent')
ax1.tick_params(axis='y', labelcolor='blue')
plt.xticks(rotation=90)
# Create a second y-axis for 'TotalCases' data
ax2 = ax1.twinx()
sns.barplot(data=data, y='TotalCases', x='Continent', ci=True, color='orange', ax=ax2)
ax2.set_ylabel('Total Cases', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

plt.xticks(rotation=90)
plt.show()
```

```python
grouped_data = data.groupby('Country')

# Iterate over each country and create plots
for country, country_data in grouped_data:
    plt.figure(figsize=(8, 6))

    # Create a bar plot for 'Population' data
    ax1 = sns.barplot(data=country_data, y='Population', x='Continent', ci=True, color='blue')
    ax1.set_ylabel('Population', color='blue')
    ax1.set_xlabel('Continent')
    ax1.set_title(f'Population and Total Cases in {country}')
    ax1.tick_params(axis='y', labelcolor='blue')

    # Create a second y-axis for 'TotalCases' data
    ax2 = ax1.twinx()
    sns.barplot(data=country_data, y='TotalCases', x='Continent', ci=True, color='green', ax=ax2)
    ax2.set_ylabel('Total Cases', color='orange')
    ax2.tick_params(axis='y', labelcolor='orange')

    plt.xticks(rotation=90)
    plt.show()
```
